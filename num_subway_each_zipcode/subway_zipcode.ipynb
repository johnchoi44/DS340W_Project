```python
import geopandas as gpd
import pandas as pd
import numpy as np
from shapely.geometry import Point

subway_df = pd.read_csv('/Users/dnylee/DS340W/Dataset/MTA_Subway_Stations.csv')

nyc_zipcode = gpd.read_file('/Users/dnylee/DS340W/DS340W_Project/Modified Zip Code Tabulation Areas (MODZCTA)_20241012.geojson')

# Create geometry for each subway station
subway_df['geometry'] = subway_df.apply(lambda row: Point(float(row['GTFS Longitude']), float(row['GTFS Latitude'])), axis=1)

subway_gdf = gpd.GeoDataFrame(subway_df, geometry='geometry', crs="EPSG:4326")

# Perform spatial join with ZIP code GeoDataFrame using the 'predicate' argument
subway_gdf_zip = gpd.sjoin(subway_gdf, nyc_zipcode, how='left', predicate='within')

subway_gdf_zip['zipcode'] = subway_gdf_zip['modzcta']

subway_final = subway_gdf_zip.drop(columns=['index_right', 'modzcta'])

print(subway_final.head())

subway_final.to_csv('/Users/dnylee/DS340W/Dataset/Subway_Station_zipcode.csv', index=False)
```
