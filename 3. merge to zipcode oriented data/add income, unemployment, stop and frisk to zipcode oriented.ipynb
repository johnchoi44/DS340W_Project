## merge income to zipcode oriented data

### load income data

```python
from google.colab import drive
drive.mount('/content/drive')

df_income = pd.read_csv('/content/drive/MyDrive/ds340/nypd complaint data/3. merge to zipcode oriented data/2022_5yrs_income.csv')

df_income.head()
```

### preprocess income data


```python
df_income['Geographic Area Name'] = df_income['Geographic Area Name'].str.replace('ZCTA5', '', regex=True)
df_income.rename(columns={'Geographic Area Name': 'zipcode'}, inplace=True)
df_income['zipcode'] = df_income['zipcode'].astype(str)
df_income['zipcode'] = df_income['zipcode'].str.strip()
df_income.rename(columns={'Estimate!!Households!!Median income (dollars)': 'median_income'}, inplace=True)
df_income.head()
```

### load zipcode oriented data


```python
df_zipcode = pd.read_csv('/content/drive/MyDrive/ds340/nypd complaint data/2. covert to zipcode oriented/zipcode_oriented.csv')

df_zipcode.head()
```

### preprocess zipcode oriented data

```python
df_zipcode['zipcode'] = df_zipcode['zipcode'].astype(str)
df_zipcode['zipcode'] = df_zipcode['zipcode'].str.replace(r'\.0$', '', regex=True)
df_zipcode['zipcode'] = df_zipcode['zipcode'].str.strip()
df_zipcode.head()
```

### merge

```python
df_zipcode = df_zipcode.merge(df_income[['zipcode','median_income']], on='zipcode', how='left')
df_zipcode
```

## preprocess final data

```python
df_zipcode.info()
```

```python
df_zipcode
```

```python
df_zipcode.isnull().sum()
```

```python
df_zipcode.rename(columns = {'Unnamed: 0': 'index'}, inplace=True)
```

```python
df_zipcode = df_zipcode[df_zipcode['zipcode'] != '99999']
df_zipcode = df_zipcode[df_zipcode['zipcode'] != 'nan']
```

```python
df_zipcode['median_income'] = df_zipcode['median_income'].str.replace('+', '', regex=False)
df_zipcode['median_income'] = df_zipcode['median_income'].str.replace(',', '', regex=False)

df_zipcode['num_stopandfrisk'] = df_zipcode['num_stopandfrisk'].fillna('0')

df_zipcode['num_police_station'] = df_zipcode['num_police_station'].fillna('0')

df_zipcode['num_restroom'] = df_zipcode['num_restroom'].fillna('0')

df_zipcode['num_subway_stations'] = df_zipcode['num_subway_stations'].fillna('0')

df_zipcode['num_bus_stops'] = df_zipcode['num_bus_stops'].fillna('0')
```

```python
df_zipcode = df_zipcode.astype({
    'zipcode': str,
    'num_felony': int,
    'num_violation': int,
    'num_misdemeanor': int,
    'most_freq_ofns_desc': str,
    'most_freq_prem_type': str,
    'num_total_crimes': int,
    'median_income': float,
    'unemployment_rate': float,
    'num_stopandfrisk': int,
    'num_police_station': int,
    'num_restroom': int,
    'num_bus_stops': int,
    'num_subway_stations': int,
    'pop_est': int
})
```
