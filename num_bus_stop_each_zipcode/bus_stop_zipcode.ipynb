```python
import geopandas as gpd
import pandas as pd
import numpy as np
from shapely.geometry import Point

df = pd.read_csv('/Users/dnylee/DS340W/Dataset/Bus_Stop_Shelter_20241107.csv')

nyc_zipcode = gpd.read_file('/Users/dnylee/DS340W/DS340W_Project/Modified Zip Code Tabulation Areas (MODZCTA)_20241012.geojson')

# Create geometry for each bus stop
df['geometry'] = df.apply(lambda row: Point(float(row['Longitude']), float(row['Latitude'])), axis=1)

df_gdf = gpd.GeoDataFrame(df, geometry='geometry', crs="EPSG:4326")

# Perform spatial join with ZIP code GeoDataFrame using the 'predicate' argument
df_gdf_zip = gpd.sjoin(df_gdf, nyc_zipcode, how='left', predicate='within')

df_gdf_zip['zipcode'] = df_gdf_zip['modzcta']

df_final = df_gdf_zip.drop(columns=['index_right', 'modzcta'])

print(df_final.head())

df_final.to_csv('/Users/dnylee/DS340W/Dataset/Bus_Stop_Shelter_zipcode.csv')
```
